### Ejercicio 1: Crea una lista con los múltiplos de 7 entre 1 y 100 usando comprehension.

### Ejercicio 2: Dado el diccionario datos = {'a':10, 'b':25, 'c':7, 'd':18}
###a) Crea una lista con las claves cuyo valor sea mayor a 12.
###b) Crea un nuevo diccionario con esas claves y sus valores al cuadrado

```python
datos = {'a': 10,
         'b': 25,
         'c': 7,
         'd': 18}

# a) Lista de claves con valor > 12
Claves_mayor12 = [k for k, v in datos.items() if v > 12]

# b) Diccionario con valores al cuadrado
datosCuadrado = {k: v**2 for k, v in datos.items() if v > 12}

print(f"Claves seleccionadas: {Claves_mayor12}")
print(f"Nuevo diccionario: {datosCuadrado}")
```

###Ejercicio 3: Define una función 'resumen(lista)' que retorne un diccionario con:
### longitud, suma, promedio, mínimo y máximo de la lista.

```python
def resumen(lista):
    if not lista:
        return "Error: No se proporcionaron datos para el análisis."

    # La función realiza el análisis dinámico de la lista
    return {
        'longitud': len(lista),
        'suma': sum(lista),
        'promedio': sum(lista) / len(lista),
        'minimo': min(lista),
        'maximo': max(lista)
    }


mis_datos =[1,2,3,8,4,5,6]

resultado = resumen(mis_datos)

print("Analisis")

if isinstance(resultado, dict):
    for clave, valor in resultado.items():

        if clave == 'promedio':
            print(f"{clave.capitalize()}: {valor:.2f}")
        else:
            print(f"{clave.capitalize()}: {valor}")
```

### Ejercicio 4: Usa map() para aplicar la función potencia(x, 3) a la lista [1,2,3,4,5].

```python

def potencia(base, exponente=2):
    return base ** exponente

numeros = [1,2,3,4,5]

cubos = list(map(lambda x: potencia(x, 3), numeros))

print(f"Originales: {numeros}")
print(f"Resultados (al cubo): {cubos}")
```

### Ejercicio 1: Dado A = np.arange(1, 26).reshape(5, 5)
###a) Extrae la sub-matriz de las 3 últimas filas y las 3 primeras columnas.
###b) Extrae la segunda columna completa.
###c) Extrae los elementos de los ángulos (esquinas) de la matriz.

```python
import numpy as np

A = np.arange(1, 26).reshape(5, 5)
print("Matriz A:\n", A)

# a) 3 últimas filas y 3 primeras columnas
sub_matriz = A[2:, :3]
print("\na) Submatriz:\n", sub_matriz)

# b) Segunda columna completa
segunda_columna = A[:, 1]
print("\nb) Segunda columna:\n", segunda_columna)

# c) Elementos de las esquinas
esquinas = [A[0,0], A[0,-1], A[-1,0], A[-1,-1]]
print("\nc) Esquinas:\n", esquinas)
```

###Ejercicio 2: Con la misma matriz A:
###a) Crea una máscara booleana para los valores mayores a 12.
###b) Reemplaza esos valores por -1 en una copia del array.
###c) Usa np.where para reemplazar impares por 0 y mantener los pares.

```python
# a) Máscara booleana (>12)
mascara = A > 12
print("\na) Máscara:\n", mascara)

# b) Reemplazar >12 por -1 en copia
B = A.copy()
B[mascara] = -1
print("\nb) Matriz modificada:\n", B)

# c) Reemplazar impares por 0 (pares se mantienen)
C = np.where(A % 2 == 0, A, 0)
print("\nc) Pares se mantienen, impares = 0:\n", C)
```

### Ejercicio 1: Crea una matriz 5×4 de enteros aleatorios entre 1 y 100 (seed=7).
### a) Calcula la media, máximo y mínimo de cada columna.
### b) Calcula la suma acumulada de la primera fila.
### c) Normaliza la matriz para que cada columna tenga valores entre 0 y 1.
### Fórmula: (x - min_col) / (max_col - min_col)

```python
import numpy as np

np.random.seed(7)
M = np.random.randint(1, 101, (5, 4))

print("Matriz:\n", M)

# a) media, max y min por columna
media = M.mean(axis=0)
maximos = M.max(axis=0)
minimos = M.min(axis=0)

print("\na) Media:", media)
print("Máximos:", maximos)
print("Mínimos:", minimos)

# b) suma acumulada de la primera fila
acum = np.cumsum(M[0])
print("\nb) Suma acumulada primera fila:", acum)

# c) normalización por columna
norm = (M - minimos) / (maximos - minimos)
print("\nc) Matriz normalizada:\n", norm.round(2))
```

###Ejercicio 3: Con el array de notas y nombres del ejemplo de argsort:
###a) ¿Cuántos estudiantes tienen nota mayor al promedio? (usa máscara booleana)
### b) Imprime solo los nombres de esos estudiantes.

```python
nombres = np.array(['Carlos','Ana','Diana','Pedro','Eva','Luis'])
notas = np.array([75, 92, 88, 61, 95, 78])

# a) promedio
prom = notas.mean()
mask = notas > prom

print("\na) Promedio:", prom)
print("Cantidad mayor al promedio:", np.sum(mask))

# b) nombres de esos estudiantes
print("\nb) Estudiantes con nota mayor al promedio:")
print(nombres[mask])
```

###Ejercicio 4: Crea dos matrices 3×4 y concaténalas vertical y horizontalmente.
### Luego divide el resultado vertical en 3 partes iguales con vsplit.

```python
A = np.random.randint(1, 10, (3, 4))
B = np.random.randint(1, 10, (3, 4))

print("\nMatriz A:\n", A)
print("\nMatriz B:\n", B)

# concatenación vertical
vertical = np.concatenate([A, B], axis=0)
print("\nVertical shape:", vertical.shape)

# concatenación horizontal
horizontal = np.concatenate([A, B], axis=1)
print("Horizontal shape:", horizontal.shape)

# dividir vertical en 3 partes
partes = np.vsplit(vertical, 3)

print("\nPartes:")
for i, p in enumerate(partes):
    print(f"Parte {i+1}:\n", p)
```
